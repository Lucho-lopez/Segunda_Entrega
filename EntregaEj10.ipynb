1. El generar_lista_alumnos que recibe como paramentro una lista vacia, abre los archivos.txt en modo lectura, luego utilizo el csv.reader para hacer una separacion de los archivos por comas por lo cual los elementos del archivo quedarian separados en diferentes listas, despues recorro todo el archivo de los nombres de los estudiantes, por cada iteracion cargo en una lista agregando al final de la misma el nombre del estudiante con la respectiva suma de sus notas, lo que hago con las notas es acceder al dato del archivo de las notas en la posicion en la que se encuentra esta y la convierto en un entero para poder asi hacer la suma de las mismas sin necesidad de crear otras variables, y por ultimo una vez ya terminado el recorrido, cierro los archivos ya que no los voy a necesitar mas.

```python
import csv

def generar_lista_alumnos (lista):
    eval1 = open('eval1.txt', 'r')
    eval2 = open('eval2.txt', 'r')
    nombre = open ('nombres_1.txt', encoding='utf-8')
    notas1 = csv.reader(eval1, delimiter=',')
    notas2 = csv.reader(eval2, delimiter=',')
    nombres = csv.reader(nombre, delimiter=',')
    
    for i in nombres:
        lista.append([i[0], (int((next(notas1))[0])) + (int((next(notas2))[0]))])
    eval1.close()
    eval2.close()
    nombre.close()
```

2. El calcular_promedio que recibe como parametro una lista con los datos cargados, inicializo una variable total en 0 para utilizarla como acumulador, luego recorro la lista sumando todas las notas de los alumnos, una vez terminada la iteracion por todos los alumnos esta funcion retorna el promedio (el total de las notas dividido la cantidad de alumnos).

```python
def calcular_promedio (lista):
    total = 0
    for alumno in lista:
        total = total + alumno[1]
    return (total // len(lista))
```

3. El informar_alumnos_debajo_promedio que recibe como parametro una lista con los datos cargados, creo la variable promedio y le asigno el retorno de la funcion calcular_promedio explicada en el punto 2, luego recorro la lista de alumnos revisando quienes estan por debajo del promedio, si el alumno esta por debajo del promedio informa su nombre y la nota que saco, si esta en el promedio o por encima no pasa nada.

```python
def informar_alumnos_debajo_promedio (lista):
    promedio = calcular_promedio(lista)
    print (f'Los alumnos que obtuvieron notas por debajo del promedio ({promedio}) son: ')
    for alumno in lista:
        if (alumno[1] < promedio):
            print (f"nombre: {alumno[0]}, nota: {alumno[1]}")
```
